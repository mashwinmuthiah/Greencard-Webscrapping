```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
url = 'https://www.dhs.gov/immigration-statistics/yearbook/2017/table2'
html=urlopen(url).read()
soup=BeautifulSoup(html,'html.parser')
```

```python
table2 = {
    'Region and country of last residence':[],
    '2015':[],
    '2016':[],
    '2017':[]
    
}
```

```python
table2
```

```python
tables = soup.find_all('table')
rows = tables[0].findAll('tr')[1:]
```

```python
for r in rows:
    r1=r.findAll('th')
    r2=r.findAll('td')
    table2['Region and country of last residence'].append(r1[0].get_text())
    table2['2015'].append(r2[0].string)
    table2['2016'].append(r2[1].string)
    table2['2017'].append(r2[2].string)
    
    

        
```

```python
table2
```

```python
df=pd.DataFrame(table2)
```

```python
df.to_csv('Table2.csv',index=False)
```
