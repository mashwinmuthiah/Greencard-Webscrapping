```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url='https://www.dhs.gov/immigration-statistics/yearbook/2017/table3'
html=urlopen(url).read()
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
table3={
    'Region of Birth':[],
    '2015':[],
    '2016':[],
    '2017':[]
}
```

```python
tables = soup.find_all('table')
rows = tables[0].find_all('tr')[1:]
```

```python
for r in rows:
    col1=r.find_all('th')
    col2=r.find_all('td')
    table3['Region of Birth'].append(col1[0].get_text())
    table3['2015'].append(col2[0].string)
    table3['2016'].append(col2[0].string)
    table3['2017'].append(col2[0].string)
    
```
