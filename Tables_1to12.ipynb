```python
tables = soup[0].find_all('table')
rows = tables[0].findAll('tr')[1:]

for row in rows:
    cols = row.findAll('td')
    row1 = row.findAll('th')
    ds_table1['number'].append(cols[0].string)
    ds_table1['year'].append(row1[0].string)
df = pd.DataFrame(ds_table1)
df.to_csv('Table 1. Persons Obtaining Lawful Permanent Resident Status.csv', index=False)
```
